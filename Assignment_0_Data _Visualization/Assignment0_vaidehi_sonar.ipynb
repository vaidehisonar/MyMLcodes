## II. REVIEW

### A. LINEAR ALGEBRA

Linear algebra is a form of continuous mathematics. It forms the basis of machine learning algorithms and especially deep learning algorithms. The study of linear algebra consists of the following mathematical objects:<br/>
**Scalars**: It is a single number.<br/>
**Vectors**: Vector is an array of numbers.<br/>
**Matrices**:A matix is a 2-D array of numbers, where element is identified by two indices representing rows and columns.<br/>
**Tensors**: An array of numbers in a grid with variable number of axes.<br/>

**Transpose**: One important operation on matrices is the transpose. The transpose of a matrix is the mirror image of the matrix across the main diagonal line. Transpose is denoted as,
$\displaystyle 
(A^T)i,j= Aj,i.$

**Matrix Multiplication**: The matrix product of matrices A and B is a third matrix C,where A must have the same number of columns as B has rows. If A is of shape m × n and B is of shapen × p, then C is of shape m × p.<br/>
                                    **C=AB**<br/>
The product operation is deﬁned by,<br/>
$\displaystyle 
                                Ci,j= \sum_{n=1}^{\infty}Ai,kBk,j$
                                
Matrix multiplication is distributive,associative and commutative.

**Identity Matrix**: An identity matrix is a matrix that does not change any vector when the vector is multiplied by that matrix. Identity matrix that preserves n-dimensional vector is denoted as $\displaystyle I_n $.

                     
**Span**: The span of a set of vectors is the set of all points obtainable by linear combinationof the original vectors.

**Norm**: In machine learning, the size of vectors is measured using a function called a norm.

**Different Types of Matrices**:
1) Diagonal Matrix: Non-zero entries along diagonal of the matrix and other entries as zeros.<br/>
2) Symmetric Matrix: A  matrix that is eual to its transpose.<br/>
3) Orthogonal Matrix: An orthogonal matrix is a square matrix whose rows and columns are mutually orthonormal among themselves.
  $\displaystyle  A^TA = AA^T = I $<br/>
  
**Eigendecomposition**: It is type of matrix decomposition in which a matrix is decomposed into a set of eigenvectors and eigenvalues. An eigenvector of a square matrix A is a nonzero vector v such that multiplication by A alters only the scale of v<br/>
 $\displaystyle Av = \lambda v$<br/>
 where, the scalar $\displaystyle \lambda $ is called the eigenvalue.
 
**Trace Operator**: Trace operator gives the sum of all diagonal entries of a matrix.

**Determinant of a matrix**: It id denoted by det(A). Determinant is eual to the product of all the eigenvalues of the matrix. The absolute value of the determinant is a measure of how much multiplication by the matrix expands or contracts space. 



### B. PROBABILITY

Probability is a branch of mathematics which calculates the likelihood of an event that will occur. It  is expressed as a number between 0 and 1. Probability represents uncertainities and is therefore and important part of machine learning.

**Random variables**: A random variable is a variable that can have any random value.It is discrete or continuous.

**Probability Distribution**: A probability distribution is a distribution of likelihood of a random variable to take on each of its possible states.

**Probability Mass Function**: A probability distribution over discrete variables may be described using a proba-bility mass function(PMF) and is usually denoted by P.

**Continuous Variables and Probability Density Function**: ProbabilP(y=*y*/x=*x*)ity density function is a function of continuous random variable, whose integral across an interval gives the probability that the value of the variable lies within the same interval.<br/>
Probability of x lying in the interval a to b is given by:

$$\begin{align} \Pr [a \le X \le b] = \int_a^b f_X(x) \, dx \end{align}$$

**Marginal Probability**: Marginal distribution of a subset of a collection of random variables is the probability distribution of the variables contained in the subset.<br/>

Sum rule:$$\begin{align} \Pr(X=x) = \sum_y \Pr(X=x,Y=y) \end{align}$$

**Conditional Probability**: It is the probability of an event given some another event has occurred.<br/>

**Bernoulli Distribution**: It is a distribution that is controlled by a single paramenter.

**Multinoulli Distribution**: Multinoulli Distribution is a districution over a single discrete variable with k finite states.

**Gaussian Distribution**: It is known as normal distribution which is used for real numbers.

**Bayes Rule**: It describes the probability of an event, based on prior knowledge of conditions that might be related to the event.
$$\begin{align} P(A\mid B) = \frac{P(B \mid A) \, P(A)}{P(B)}\end{align}$$

- Waterfront houses does not have any relevance in price prediction. Although prices for view and waterfront houses seem to overlap.

#### Bar Plot:
